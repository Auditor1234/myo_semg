```python
import numpy as np
import matplotlib.pyplot as plt
from data_process import load_movs_from_file, temporal_normal

subject = 1
mat_file = 'D:/Download/Datasets/Ninapro/DB2/DB2_s%d/S%d_E1_A1.mat' % (subject, subject)
movements, label = load_movs_from_file(mat_file)

# original signal
channels = 8
emg_type = np.array(movements[1]) * 10000
print(len(emg_type), 'points')
for i in range(channels):
    plt.subplot(8, 1, i + 1)
    plt.plot(emg_type[:, i])
    plt.axis('off')
plt.show()

plt.clf()
mov_len = len(emg_type)
emg_type = emg_type[int(0.25 * mov_len) : int(0.75 * mov_len)]
for i in range(channels):
    plt.subplot(8, 1, i + 1)
    plt.plot(emg_type[:, i])
    plt.axis('off')
plt.show()
```
